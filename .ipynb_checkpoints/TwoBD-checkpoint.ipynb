Загрузим датасет "Открытые бассейны" 
https://data.mos.ru/opendata/7708308010-basseyny-plavatelnye-otkrytye

```python
dataset_id = 894
url =  f'https://apidata.mos.ru/v1/datasets/{dataset_id}/rows?api_key={api_key}'
print(url)
response = requests.get(url)   
pools2 = response.json() 
```

Объединим два датасета по районам

```python
from itertools import chain
def find_all_for_district(district):
    entertainments = list(filter(lambda entert: entert['Cells']['District'] == district,chain(pools,pools2)))
    return entertainments
    
districts ={entert['Cells']['District'] for entert in  chain(pools,pools2)}
resulted_data  = {district:find_all_for_district(district) for district in districts}
resulted_data
```

```python
resulted_data.keys()
```

```python
import json

with open('rayon.json','w+',encoding='utf-8') as f:
    json.dump(resulted_data,f)
```
