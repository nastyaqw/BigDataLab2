```python
import time
import json
from dateutil.relativedelta import *
from datetime import date

import vk_requests
```

```python
with open('vk_data.json','r') as f:
    vk_data = json.load(f)
api = vk_requests.create_api(service_token=vk_data['token'])
```

```python
#имя группы из url
display_name = 'becoart'
group_info = api.groups.getById(group_ids=display_name)
group_info
```

Посмотрим данные о пользователях

```python
group_id = group_info[0]['id']
response = api.groups.getMembers(group_id=group_id,fields='bdate')
response
```

Расчитаем возраст

```python
def is_valid(member):
    return len(member.get('bdate','').split('.')) > 2

def calculate_age(bday):
    day,month,year = map(int,bday.split('.'))
    born = date(day=day,month=month,year=year)
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

members = response['items']
valid=[{'id':member['id'],'age':calculate_age(member['bdate'])} for member in members if is_valid(member)]
valid
```
